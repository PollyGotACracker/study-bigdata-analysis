```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sb
```

```python
x=np.linspace(0,14,100)
y1=np.sin(x)
y2=2*np.sin(x+0.5)
y3=3*np.sin(x+1.0)
y4=4*np.sin(x+1.5)

plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
```

```python
sb.set_style('whitegrid')

plt.figure(figsize=(10,6))
plt.plot(x,y1,x,y2,x,y3,x,y4)
plt.show()
```

```python
#tips :           
tips=sb.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(8,6))
sb.boxplot(x='day',y='total_bill', data=tips)
plt.show()
```

```python
#hue             .
plt.figure(figsize=(8,6))
sb.boxplot(x='day',y='total_bill', hue='smoker',data=tips , palette='Set3')
plt.show()
```

```python
sb.set_style('darkgrid')
sb.lmplot(x='total_bill', y='tip',data=tips,size=7)
plt.show()
```

```python
sb.lmplot(x='total_bill', y='tip',hue='smoker',data=tips,palette='Set1',size=7)
plt.show()
```

```python
flights=sb.load_dataset('flights')
flights.head(5)
```

```python
#      
flights=flights.pivot("month","year","passengers")
flights.head(5)
```

```python
plt.figure(figsize=(10,8))
sb.heatmap(flights, annot=True, fmt='d') 
#annot :  True      fmt :     .2f  
plt.show()
```
